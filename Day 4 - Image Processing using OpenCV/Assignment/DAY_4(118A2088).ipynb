```python
img1=cv2.imread("Color.jpeg",cv2.IMREAD_GRAYSCALE)

img2 = np.zeros((img1.shape[0], img1.shape[1]),dtype=np.uint8)
cv2.randn(img2, 128, 20) 
cv2_imshow(img2)
cv2.imwrite("Gaussian random noise.jpg",img2)
```

```python
uniform_noise = np.zeros((img1.shape[0], img1.shape[1]),dtype=np.uint8)
cv2.randu(uniform_noise,0,255)
img3 = uniform_noise.copy()
ret,imgn3= cv2.threshold(uniform_noise,250,255,cv2.THRESH_BINARY)
cv2_imshow(img3)
cv2.imwrite("Impuls noise.jpg",img3)
```

average filtering 


*   1.image with gaussian noise




```python
img2 = (img2*0.1).astype(np.uint8)
imgn2 = cv2.add(img1,img2)
processed_image2 = cv2.blur(imgn2,(5,5))
st2=np.hstack((imgn2,processed_image2))
cv2_imshow(st2)


```

average filtering 
 

*  2.image with impulse noise







 



```python
img3 = (img3*0.5).astype(np.uint8)
imgn3 = cv2.add(img1,img3)

processed_image3 = cv2.blur(imgn3,(10,5))
st3=np.hstack((imgn3,processed_image3))
cv2_imshow(st3)
```

gaussian filtering

*   1.image with gaussian noise




```python
image_gaussian_processed = cv2.GaussianBlur(imgn2,(5,5),1)
gst2=np.hstack((imgn2,processed_image2,image_gaussian_processed))
cv2_imshow(gst2)
```

gaussian filtering



* 2.image with impulse noise






```python
image_impulse_processed = cv2.GaussianBlur(imgn3,(5,5),1)
gst3=np.hstack((imgn2,processed_image3,image_impulse_processed))
cv2_imshow(gst3)
```

Median Filtering

*   1.image with gaussian noise




```python
blurred2 = cv2.medianBlur(imgn2, 3)
mst2=np.hstack((imgn2,blurred2))
cv2_imshow(mst2)
cv2.imwrite("Median filter - Gaussian noise.jpg",blurred2)
```



*  2.image with impulse noise




```python
blurred3 = cv2.medianBlur(imgn3, 3)
mst3=np.hstack((imgn3,blurred3))
cv2_imshow(mst3)
cv2.imwrite("Median filter - Impulse noise.jpg",blurred3)
```

Bilateral Filtering

*  1.image with gaussian noise




```python
bilateral2 = cv2.bilateralFilter(imgn2, 5, 75, 75)
bst2=np.hstack((processed_image2,image_gaussian_processed,bilateral2))
cv2_imshow(bst2)
```



*  2.image with impulse noise




```python
bilateral3 = cv2.bilateralFilter(imgn3, 5, 75, 75)
bst3=np.hstack((processed_image3,image_impulse_processed,bilateral3))
cv2_imshow(bst3)
```
